```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('test_visualization_nolinks.csv')
```

```python
data
```

```python
# plotting monthly rent for each apartment
plt.plot(data.price,'.')
plt.title("Scraped apartment data: rent")
plt.xlabel("Apartment ID")
plt.ylabel("rent [$/month]")
plt.show()

print("Average rent: "+ str(data.price.mean()))
```

```python
# plotting monthly rent for each apartment, in range $[0 5000]

plt.plot(data.price,'.')
plt.title("Scraped apartment data: rent in range[0,5000]")
plt.xlabel("Apartment ID")
plt.ylabel("rent [$/month]")
plt.ylim((0,5000))
plt.show()

print("Average rent: "+ str(data[data.price <= 5000].price.mean()))
```

```python
# Above plot, adjusted for # of beds
data_adjusted4beds = data.price/data.beds
plt.plot(data_adjusted4beds,'.')
plt.title("Monthly rent adjusted for #beds, in range [0 5000]")
plt.xlabel("Apartment ID")
plt.ylabel("rent per bedroom [$/month]")
plt.ylim((0,5000))
plt.show()

print("Average rent (adjusted for #bedrooms): "+ str(data_adjusted4beds[data.price <= 5000].mean()))
```

```python
#histogram with 50 bins
plt.hist(data_adjusted4beds, bins=50)
plt.title("Histogram of monthly rent per apartment")
plt.xlabel("Rent [$/month]")
plt.ylabel("# of apartments")
plt.show()
```

```python
#histogram in range [0 10000] bin size 100
plt.hist(data_adjusted4beds, bins=range(0,10000,100))
plt.title("Histogram of monthly rent per apartment")
plt.xlabel("Rent [$/month]")
plt.ylabel("# of apartments")
plt.show()
```

```python
# price over squarefootage
plt.plot(data.ft2,data.price,'.')
plt.xlim((0,2000))
plt.title("price per ft2")
plt.xlabel("size [ft2]")
plt.ylabel("Rent [$/month]")
plt.show()
```

```python
# price per squarefootage

print((data.price/data.ft2).mean())
plt.plot(data.price/data.ft2,'.')
plt.ylim((0,10))
plt.title("$/ft2 for each apartment")
plt.xlabel("Apartment ID")
plt.ylabel("[$/ft2]")
plt.show()

mean_ppsf = (data.price/data.ft2).mean()
print("Average $/ft2: "+ str(mean_ppsf))
```
